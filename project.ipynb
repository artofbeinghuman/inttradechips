```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.rcParams['figure.figsize'] = [14, 8]
```

```python
def trans2num(num):
    num = num[1:]
    if num[-1] == 'B':
        num = float(num[:-1])
    elif num[-1] == 'M':
        num = float(num[:-1])/10**3
    else:
        print(num)
    return num
```

```python
# plot absolute trade

gp = pd.read_csv('GER2TAI19_panel95-19.csv')
tp = pd.read_csv('TAI2GER_panel95-19.csv')
y = np.array(gp[:25]['Year'])
                 
for i in range(len(gp['Current Gross Export'])):
    gp.loc[i,'Current Gross Export'] = trans2num(gp.loc[i,'Current Gross Export'])

for i in range(len(tp['Current Gross Export'])):
    tp.loc[i,'Current Gross Export'] = trans2num(tp.loc[i,'Current Gross Export'])

gm = np.array(gp.loc[gp['Sector']=='Machinery','Current Gross Export'])
ge = np.array(gp.loc[gp['Sector']=='Electronics','Current Gross Export'])
te = np.array(tp.loc[tp['Sector']=='Electronics','Current Gross Export'])
tm = np.array(tp.loc[tp['Sector']=='Machinery','Current Gross Export'])


plt.plot(y,tm,label='TAI->GER Machinery', color='red',linestyle='dashed')
plt.plot(y,gm,label='GER->TAI Machinery', color='red')
plt.plot(y,te,label='TAI->GER Electronics', color='blue',linestyle='dashed')
plt.plot(y,ge,label='GER->TAI Electronics', color='blue')
plt.xlim(left=1995,right=2019)
plt.legend()
plt.ylabel('Exports in billion USD')
plt.xlabel('Year')
plt.show()
```

```python
# plot change in trade

gp = pd.read_csv('GER2TAI19_panel95-19.csv')
tp = pd.read_csv('TAI2GER_panel95-19.csv')
y = np.array(gp[:25]['Year'])
                 
for i in range(len(gp['Current Gross Export'])):
    gp.loc[i,'Current Gross Export'] = trans2num(gp.loc[i,'Current Gross Export'])

for i in range(len(tp['Current Gross Export'])):
    tp.loc[i,'Current Gross Export'] = trans2num(tp.loc[i,'Current Gross Export'])
    
gm = np.array(gp.loc[gp['Sector']=='Machinery','Current Gross Export'])
ge = np.array(gp.loc[gp['Sector']=='Electronics','Current Gross Export'])
te = np.array(tp.loc[tp['Sector']=='Electronics','Current Gross Export'])
tm = np.array(tp.loc[tp['Sector']=='Machinery','Current Gross Export'])

gm = gm/gm[0]*100
ge = ge/ge[0]*100
tm = tm/tm[0]*100
te = te/te[0]*100

plt.plot(y,tm,label='TAI->GER Machinery', color='red',linestyle='dashed')
plt.plot(y,gm,label='GER->TAI Machinery', color='red')
plt.plot(y,te,label='TAI->GER Electronics', color='blue',linestyle='dashed')
plt.plot(y,ge,label='GER->TAI Electronics', color='blue')
plt.hlines(100, 1990, 2025, linestyle='dotted')
plt.xlim(left=1995,right=2019)
plt.legend()
plt.ylabel('Change in exports in %')
plt.xlabel('Year')
plt.show()
```
